## groupby

```python
# data from: 
ipl_data = {'Team': ['Riders', 'Riders', 'Devils', 'Devils', 'Kings',
         'kings', 'Kings', 'Kings', 'Riders', 'Royals', 'Royals', 'Riders'],
         'Rank': [1, 2, 2, 3, 3,4 ,1 ,1,2 , 4,1,2],
         'Year': [2014,2015,2014,2015,2014,2015,2016,2017,2016,2014,2015,2017],
         'Points':[876,789,863,673,741,812,756,788,694,701,804,690]}

df = pd.DataFrame(ipl_data)
df
```

```python
df.groupby("Team")["Points"].sum() # 팀마다 포인트가 얼마나 쌓였나
```

```python
h_index = df.groupby(["Team", "Year"])["Points"].sum()
h_index
```

```python
h_index["Devils":"Kings"]
```

```python
h_index.unstack() #Goup으로 묶여진 데이터를 matrix 형태로 바꿔줌.
```

```python
h_index.swaplevel() #index level변경 가능
```

```python
h_index.sum(level=0) #team 별로 합산
```

```python
h_index.sum(level=1) #년도 별로 합산
```

## pivot table

```python
import dateutil
```

```python
df_phone = pd.read_csv("file:///C:/Users/call2/Downloads/phone_data.csv")
df_phone
```

```python
df_phone['date'] = df_phone['date'].apply(dateutil.parser.parse, dayfirst=True)
df_phone.head() #dateutil : datautil 데이타 형 변환(시간 날짜...)
```

```python
df_phone.pivot_table(["duration"],
                     index=[df_phone.month,df_phone.item], 
                     columns=df_phone.network, aggfunc="sum", fill_value=0)
```

```python
df_phone[df_phone['item'] == 'call'].groupby('month')['duration'].sum()
# item이 call인 경우의 index만 월을 기중으로 duration의 sum
```

```python
df_phone.groupby(['month','item']).agg({'duration':sum,'network_type':"count",'date':'first'})
# 월 과 item을 기준으로 groupby을 한 다음 aggregtion으로 한번에 볼 수 있게 묶어줌.
```

```python
grouped = df_phone.groupby('month').agg({"duration" : [min,max]})
grouped
```

```python
df_phone.pivot_table(["duration"],index=[df_phone.month, df_phone.item],columns = df_phone.network,aggfunc="sum",fill_value =0)
# pivot
# duration : data 이름 , index : 목차 , columns : 네트워크사용(분야마다 분류됨.)&sum으로 정리
```

## crosstab

```python
df_movie = pd.read_csv("https://raw.githubusercontent.com/TeamLab/machine_learning_from_scratch_with_python/master/code/ch5/data/movie_rating.csv")
df_movie.head()
```

```python
df_movie.pivot_table(["rating"], index=df_movie.critic, columns=df_movie.title,
                     aggfunc="sum", fill_value=0)
```

```python
pd.crosstab(index=df_movie.critic,columns=df_movie.title,values=df_movie.rating,
            aggfunc="first").fillna(0)
```

## merge

```python
raw_data = {
        'subject_id': ['1', '2', '3', '4', '5', '7', '8', '9', '10', '11'],
        'test_score': [51, 15, 15, 61, 16, 14, 15, 1, 61, 16]}
df_a = pd.DataFrame(raw_data, columns = ['subject_id', 'test_score'])
df_a
```

```python
raw_data = {
        'subject_id': ['4', '5', '6', '7', '8'],
        'first_name': ['Billy', 'Brian', 'Bran', 'Bryce', 'Betty'], 
        'last_name': ['Bonder', 'Black', 'Balwner', 'Brice', 'Btisan']}
df_b = pd.DataFrame(raw_data, columns = ['subject_id', 'first_name', 'last_name'])
df_b
```

```python
pd.merge(df_a, df_b, on='subject_id') # on에 컬럼에 이름을 넣어준다. 그러면 맞춰서 합쳐짐.(inner join으로)
```

```python
pd.merge(df_a, df_b, left_on='subject_id', right_on='subject_id') #colum을 여러개 넣고 싶을때.
```

## Concat 

```python
raw_data = {
        'subject_id': ['1', '2', '3', '4', '5'],
        'first_name': ['Alex', 'Amy', 'Allen', 'Alice', 'Ayoung'], 
        'last_name': ['Anderson', 'Ackerman', 'Ali', 'Aoni', 'Atiches']}
df_a = pd.DataFrame(raw_data, columns = ['subject_id', 'first_name', 'last_name'])
df_a
```

```python
raw_data = {
        'subject_id': ['4', '5', '6', '7', '8'],
        'first_name': ['Billy', 'Brian', 'Bran', 'Bryce', 'Betty'], 
        'last_name': ['Bonder', 'Black', 'Balwner', 'Brice', 'Btisan']}
df_b = pd.DataFrame(raw_data, columns = ['subject_id', 'first_name', 'last_name'])
df_b
```

```python
df_new = pd.concat([df_a, df_b])
df_new.reset_index() # 두 프레임이 합쳐짐.
```

```python
df_new = pd.concat([df_a, df_b], axis=1)
df_new.reset_index()
```
